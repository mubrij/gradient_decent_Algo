```python
class Gradient_:
    def __init__(self,data,paras):
        self.data = data
        self.paras= paras
        
    def cost_fuction(self):
        """""
        creating cost fuction
        """""
        total_cost = 0
        for i in range(len(self.data)):
            total_cost+=(1/len(self.data))*((self.data[i]*self.paras).sum()-y[i])**2
        return total_cost
    
    def gradient_descent(self,irate=0.0001,iter_value=10000):
        for i in range(iter_value):
            slope = 0
            for j in range(len(self.data)):
                for k in range(1):#rows
                    slope += ((1/len(self.data))*((self.data[j]*self.paras).sum()-y[j])*self.data[j][k])#data[j]*paras is hypothesis our prediction and - y[j] is our error
                    self.paras = self.paras-irate*slope #update
            print("cost",cost_fuction(self.data,self.paras))
            
        return slope
```

```python
class Gradient:
    def __init__(self,m_curr,b_curr,iteration,
                         n_,x,y,learning_rate):
        
        self.m_curr = m_curr
        self.b_curr= b_curr
        self.iteration=iteration
        self.n_=n_
        self.x=x
        self.y=y
        self.learning_rate=learning_rate
    
    def gradient_Algo(self):
        for i in range(self.iteration):
            
            y_pred = self.m_curr * self.x + self.b_curr
            cost = (1/self.n_) * sum([val**2 for val in (self.y - y_pred)])
            
            md = -(2/self.n_)*sum(self.x*(self.y - y_pred))
            bd = -(2/self.n_)*sum(self.y - y_pred)
                                    
            self.m_curr = self.m_curr-self.learning_rate*md
            self.b_curr = self.b_curr-self.learning_rate*bd
            print("cost",cost)
        
            
        
                
m_curr = 0
b_curr = 0
iteration =10000 
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
n_ =len(x)
learning_rate = 0.0001  

gr = Gradient(m_curr,b_curr,iteration,
              n_,x,y,learning_rate)
gr.gradient_Algo()
```
